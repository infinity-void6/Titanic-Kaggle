```python
# Embarked is converted into a matrix where rows are of length 4 and columns respectively are C S Q N
# now we transpose it to convert the columns into rows. We create specific columns in the dataframe.
# These columns will then point to the respective rows of the transposed matrix
```

```python
data.columns
```

```python
data["Sex"]=data["Sex"].replace({'male':0,'female':1})
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2)
for train_ids,test_ids in split.split(data,data[["Survived","Pclass","Sex"]]):
    strat_train_set=data.iloc[train_ids]
    strat_test_set=data.iloc[test_ids]
```

```python
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.impute import SimpleImputer

class AgeImputer(BaseEstimator,TransformerMixin):
    def fit(self,X,y=None):
        return self
    def transform(self,X):
        X=X.copy()
        imputer=SimpleImputer(strategy='mean')
        X["Age"]=imputer.fit_transform(X[["Age"]])
        return X
```

```python
from sklearn.preprocessing import OneHotEncoder

class FeatureEncoder(BaseEstimator,TransformerMixin):

    
    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        X=X.copy()
        encoder=OneHotEncoder()
        matrix_1=encoder.fit_transform(X[['Embarked']]).toarray()
        matrix_columns_1=['C','S','Q','N']
        for i in range(len(matrix_1.T)):
            X[matrix_columns_1[i]]=matrix_1.T[i]
        matrix_2=encoder.fit_transform(X[['Sex']]).toarray()
        matrix_columns_2=["M","F"]
        for i in range(len(matrix_2.T)):
            X.loc[:,matrix_columns_2[i]]=matrix_2.T[i]
        return X
```

```python
class FeatureDropper(BaseEstimator,TransformerMixin):
    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        return X.drop(["Embarked","Name","Ticket","Cabin","Sex","N"],axis=1,errors="ignore")
    
    # error = ignore is used because test data has no N column and it will rise an error
```

```python
from sklearn.pipeline import Pipeline

pipeline=Pipeline([("ageimputer",AgeImputer()),("featureencoder",FeatureEncoder()),("featuredropper",FeatureDropper())])
```

```python
strat_train_set=pipeline.fit_transform(strat_train_set)
```

```python
from sklearn.preprocessing import StandardScaler

y=strat_train_set.loc[:,"Survived"]
X=strat_train_set.drop(["Survived"],axis=1)

scaler=StandardScaler()
X_train=scaler.fit_transform(X)
y_train=y.to_numpy()
```

```python
pd.DataFrame(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

clf=RandomForestClassifier()

param_grid=[{"n_estimators":[10,100,200,500],"max_depth":[None,5,10],"min_samples_split":[2,3,4]}]

grid_search=GridSearchCV(clf,param_grid,cv=3,scoring="accuracy",return_train_score=True)
grid_search.fit(X,y)
```

```python
final_clf=grid_search.best_estimator_
```
